```python
import pandas as pd
df = pd.read_csv('titanic_train.csv')
df
```

```python
df.drop(['PassengerId','Name','Ticket','Embarked','Cabin','SibSp','Parch'],axis=1,inplace=True)
df
```

```python
target = df.Survived
inputs = df.drop('Survived',axis=1) 
```

```python
target
```

```python
inputs
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies
```

```python
inputs = pd.concat([inputs,dummies],axis=1)
inputs
```

```python
inputs.drop('Sex',axis=1,inplace=True)
inputs
```

```python
inputs.isna().any()
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs
```
